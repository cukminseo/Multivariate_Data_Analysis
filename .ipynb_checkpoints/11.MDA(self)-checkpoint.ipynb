### Q0. 데이터프레임 이름 변경

```python
# 컬럼 선택 및 이름 변경
print("Columns : ", df_raw.columns)

# 선택된 컬럼들을 새로운 리스트로 저장합니다.
selected_cols = ["조사연도", "수도권여부", "MD제공용_가구고유번호", "가구주_성별코드", "가구주_만연령", "가구원수", "가구주_교육정도_학력코드", "가구주_혼인상태코드", "자산", "부채", "처분가능소득(보완)[경상소득(보완)-비소비지출(보완)]", "가구주_산업대분류코드", "가구주_직업대분류코드", "입주형태코드"]

# 데이터프레임에서 선택된 컬럼들만 추출합니다.
df = df_raw[selected_cols]

# 컬럼 이름을 영어로 변경합니다.
df.rename(columns={
    "조사연도": "year",
    "수도권여부": "metro",
    "MD제공용_가구고유번호": "id",
    "가구주_성별코드": "sex",
    "가구주_만연령": "age",
    "가구원수": "number",
    "가구주_교육정도_학력코드": "education",
    "가구주_혼인상태코드": "marriage",
    "자산": "asset",
    "부채": "debt",
    "처분가능소득(보완)[경상소득(보완)-비소비지출(보완)]": "income",
    "가구주_산업대분류코드": "industry",
    "가구주_직업대분류코드": "job",
    "입주형태코드": "house"
}, inplace=True)

# 데이터프레임의 정보를 출력합니다.
print(df.info())

# 데이터프레임의 기술 통계를 출력합니다.
print(df.describe())
```

```python
# "income" 컬럼 추출
income = df["income"]

# "income" 컬럼의 기술 통계 출력
print(income.describe())
```

Q4. Plot the distribution of income as below with ‘seaborn’ package

```python
# ########## 시각화 ##########

# (2) 소득 분포 (박스플롯)

# 이상치 제거
income_q3 = df['income'].quantile(q=0.75)  # 3사분위수 계산
income_q1 = df['income'].quantile(q=0.25)  # 1사분위수 계산
income_iqr = income_q3 - income_q1  # IQR (사분위 범위) 계산

# 조건 설정: 소득이 3 * IQR 이하인 데이터만 선택
condition = df['income'] <= 3 * income_iqr
new_df1 = df[condition]

# 이상치를 제거한 후의 박스플롯 시각화
sns.boxplot(data=new_df1, x='income')
plt.show()
```

Q5. Randomly sample 500 households 10,000 times & draw a histogram with the averages
of each sampling data

```python
##### 500개의 가구를 무작위로 10,000번 샘플링하여 평균 소득을 구하고 히스토그램을 그립니다. #####

n = 10000  # 샘플링 횟수
mn = []  # 평균 소득을 저장할 리스트

for i in range(n):
    # 소득 데이터에서 500개의 가구를 무작위로 샘플링합니다 (복원 추출)
    sample = np.random.choice(income, size=500, replace=True)  # replace=True: 같은 값이 여러 번 선택될 수 있음
    mean = np.mean(sample)  # 샘플링된 데이터의 평균을 계산합니다
    mn.append(mean)  # 계산된 평균을 리스트에 추가합니다

# 샘플링된 평균 소득의 분포를 히스토그램으로 시각화합니다
sns.histplot(data=mn)
plt.show()
```

```python
# 99.7% 신뢰구간 계산

mean = np.mean(mn)  # 샘플링된 평균 소득의 평균
std = np.std(mn)  # 샘플링된 평균 소득의 표준 편차
SE = np.std(mn) / np.sqrt(1000)  # 표준 오차 계산

print("Mean of Incomes:", mean)
print("SE of Incomes:", SE)
print("CI with 99.7% : ", mean - (3 * SE), " ~ ", mean + (3 * SE))
```
