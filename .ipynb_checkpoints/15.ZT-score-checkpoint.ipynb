```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

##### 데이터 읽기 (kostat에서 가져옴) #####
# 2022년 가구 데이터 읽기
df_raw22 = pd.read_csv('data\\2022_household.csv', encoding='CP949')
# 2023년 가구 데이터 읽기
df_raw23 = pd.read_csv('data\\2023_household.csv', encoding='CP949')
# 판다스 옵션 설정: 모든 컬럼을 출력하도록 설정
pd.set_option('display.max_columns', None)
# 판다스 옵션 설정: 모든 행을 출력하도록 설정
pd.set_option('display.max_rows', None)
# 2023년 데이터의 처음 10개 행 출력
print(df_raw23.head(10))
```

```python
##### 컬럼 선택 및 이름 변경 #####

# 2023년 데이터의 컬럼 이름 출력
print("Columns : ", df_raw23.columns)
# 2022년 데이터의 컬럼 이름 출력
print("Columns : ", df_raw22.columns)
# 분석에 사용할 컬럼들을 선택하여 리스트로 저장
selected_cols = ["조사연도", "수도권여부", "MD제공용_가구고유번호", "가구주_성별코드", "가구주_만연령", "가구원수", "가구주_교육정도_학력코드", "가구주_혼인상태코드", "자산", "부채", "처분가능소득(보완)[경상소득(보완)-비소비지출(보완)]", "가구주_산업대분류코드", "가구주_직업대분류코드", "입주형태코드"]
# 2022년 데이터에서 선택된 컬럼들만 추출하여 새로운 데이터프레임 생성
df22 = df_raw22[selected_cols]
# 2023년 데이터에서 선택된 컬럼들만 추출하여 새로운 데이터프레임 생성
df23 = df_raw23[selected_cols]
```

```python
##### 컬럼 선택 및 이름 변경 #####

# 2022년 데이터프레임의 컬럼 이름을 영어로 변경
df22.rename(columns={
    "조사연도": "year",
    "수도권여부": "metro",
    "MD제공용_가구고유번호": "id",
    "가구주_성별코드": "sex",
    "가구주_만연령": "age",
    "가구원수": "number",
    "가구주_교육정도_학력코드": "education",
    "가구주_혼인상태코드": "marriage",
    "자산": "asset",
    "부채": "debt",
    "처분가능소득(보완)[경상소득(보완)-비소비지출(보완)]": "income",
    "가구주_산업대분류코드": "industry",
    "가구주_직업대분류코드": "job",
    "입주형태코드": "house"
}, inplace=True)

# 2023년 데이터프레임의 컬럼 이름을 영어로 변경
df23.rename(columns={
    "조사연도": "year",
    "수도권여부": "metro",
    "MD제공용_가구고유번호": "id",
    "가구주_성별코드": "sex",
    "가구주_만연령": "age",
    "가구원수": "number",
    "가구주_교육정도_학력코드": "education",
    "가구주_혼인상태코드": "marriage",
    "자산": "asset",
    "부채": "debt",
    "처분가능소득(보완)[경상소득(보완)-비소비지출(보완)]": "income",
    "가구주_산업대분류코드": "industry",
    "가구주_직업대분류코드": "job",
    "입주형태코드": "house"
}, inplace=True)

# 2022년 데이터프레임의 정보 출력
print(df22.info())

# 2023년 데이터프레임의 정보 출력
print(df23.info())
```

```python
##### 컬럼 선택 및 이름 변경 #####
# 2022년 데이터프레임에서 특정 컬럼들만 선택하여 새로운 데이터프레임 생성
df22 = df22[['year', 'id', 'income']]
# 2023년 데이터프레임에서 특정 컬럼들만 선택하여 새로운 데이터프레임 생성
df23 = df23[['year', 'id', 'age', 'sex', 'income']]
# 2022년 데이터프레임의 정보 출력
print(df22.info())
# 2023년 데이터프레임의 정보 출력
print(df23.info())
```

```python
##### 두 데이터프레임 병합 #####

# 2022년과 2023년 데이터프레임을 'id' 컬럼을 기준으로 내부 조인(inner join)하여 병합합니다.
df = pd.merge(df22, df23, on='id', how='inner')
# 병합된 데이터프레임의 정보를 출력합니다.
print(df.info())
# 병합된 데이터프레임의 마지막 3개 행을 출력합니다.
print(df.tail(3))
# 병합된 데이터프레임의 인덱스를 'id'로 설정합니다.
df1 = df.set_index('id')
# 필요한 컬럼('sex', 'age', 'income_x', 'income_y')을 선택하여 새로운 데이터프레임을 생성합니다.
df2 = df1[['sex', 'age', 'income_x', 'income_y']]
# 'income_x'와 'income_y' 컬럼 이름을 각각 'income2022'와 'income2023'으로 변경합니다.
df2.rename(columns={'income_x': 'income2022', 'income_y': 'income2023'}, inplace=True)
# 새로운 데이터프레임의 마지막 3개 행을 출력합니다.
print(df2.tail(3))
```

```python
##### 2022년과 2023년 소득의 차이를 계산 #####

# 2022년 소득과 2023년 소득의 차이를 계산하여 'diff' 컬럼에 저장합니다.
df2['diff'] = df2['income2023'] - df2['income2022']
# 새로운 데이터프레임의 정보를 출력합니다.
print(df2.info())
# 새로운 데이터프레임의 마지막 3개 행을 출력합니다.
print(df2.tail(3))
```

```python
##### 히스토그램 그리기 #####

# 그래프의 크기와 격자 설정
plt.rcParams["figure.figsize"] = (16, 4)  # 전체 그래프 크기 설정
plt.rcParams["axes.grid"] = True  # 격자 표시 설정
# 1x2 서브플롯 생성
fig, axs = plt.subplots(1, 2)
# 첫 번째 서브플롯: 2022년과 2023년 소득 차이의 밀도 그래프
axs[0].set_title("Household income difference between 2022 and 2023")  # 그래프 제목 설정
sns.kdeplot(df2['diff'], shade=False, clip=(-7500, 7500), ax=axs[0])  # 소득 차이의 커널 밀도 추정 그래프
axs[0].set_xlabel("10,000won")  # x축 레이블 설정
# 두 번째 서브플롯: 2022년과 2023년 소득 분포의 밀도 그래프
axs[1].set_title("Household income distributions in 2022 and 2023")  # 그래프 제목 설정
sns.kdeplot(df2['income2022'], shade=False, clip=(-1000, 20000), label="2022", ax=axs[1])  # 2022년 소득 분포의 커널 밀도 추정 그래프
sns.kdeplot(df2['income2023'], shade=False, clip=(-1000, 20000), label="2023", ax=axs[1])  # 2023년 소득 분포의 커널 밀도 추정 그래프
axs[1].set_xlabel("10,000won")  # x축 레이블 설정
# 범례를 오른쪽 위에 표시
plt.legend(loc="upper right")
# 그래프를 화면에 출력
plt.show()
```

```python
##### 유의성 테스트 (z-test) #####

### z-점수 계산
mn = df2['diff'].mean()  # 'diff' 컬럼의 평균 계산
std = df2['diff'].std()  # 'diff' 컬럼의 표준 편차 계산
n = len(df2)  # 데이터 개수 계산
se = std / np.sqrt(n)  # 표준 오차 계산
z_score = mn / se  # z-점수 계산

# z-점수를 소수점 한 자리까지 반올림하여 출력
print("z-score:", z_score.round(1))
```

```python
##### 유의성 테스트 (z-검정) #####

### 유의 수준 및 z-검정 계산
import scipy as sp
import scipy.stats

# 평균이 0이고 분산이 1인 정규 분포 연속 확률 변수 생성
rv = sp.stats.norm(loc=0, scale=1)

# 누적 분포 함수(CDF)를 사용하여 p-값(유의 수준)을 계산하고 출력
print("Significant Level(면적):", np.round(1 - rv.cdf(z_score), 3))  # p-값(유의 수준)을 소수점 셋째 자리까지 반올림하여 출력
```

```python
print("1.97 z스코어의 Significant Level(면적):",np.round(1-rv.cdf(1.96), 3)) # p-value(Significant level) by Cumulative distribution function
```

Exercise 2. Using the Household Financial Welfare Survey data collected from Korea National
Statistical Office, let's test whether there is a difference in income depending on the gender of
single-person household owners in their 30s.
연습 2. 통계청에서 수집한 가계금융복지조사 자료를 이용하여 30대 1인 가구주의 성별에 따라 소득에 차이가 있는지 테스트해 봅시다.

```python
##### 데이터 읽기 및 전처리 (kostat에서 가져옴) #####

# 2023년 가구 데이터를 읽어옵니다.
df_raw23 = pd.read_csv('data/2023_household.csv', encoding='CP949')
# 분석에 사용할 컬럼들을 선택하여 리스트로 저장합니다.
selected_cols = ["조사연도", "수도권여부", "MD제공용_가구고유번호", "가구주_성별코드", "가구주_만연령", "가구원수", "가구주_교육정도_학력코드", "가구주_혼인상태코드", "자산", "부채", "처분가능소득(보완)[경상소득(보완)-비소비지출(보완)]", "가구주_산업대분류코드", "가구주_직업대분류코드", "입주형태코드"]
# 선택된 컬럼들만 추출하여 새로운 데이터프레임 생성합니다.
df23 = df_raw23[selected_cols]
# 컬럼 이름을 영어로 변경합니다.
df23.rename(columns={
    "조사연도": "year",
    "수도권여부": "metro",
    "MD제공용_가구고유번호": "id",
    "가구주_성별코드": "sex",
    "가구주_만연령": "age",
    "가구원수": "number",
    "가구주_교육정도_학력코드": "education",
    "가구주_혼인상태코드": "marriage",
    "자산": "asset",
    "부채": "debt",
    "처분가능소득(보완)[경상소득(보완)-비소비지출(보완)]": "income",
    "가구주_산업대분류코드": "industry",
    "가구주_직업대분류코드": "job",
    "입주형태코드": "house"
}, inplace=True)

# 데이터프레임의 정보를 출력하여 데이터 구조와 각 컬럼의 데이터 타입, 누락 값을 확인합니다.
print(df23.info())
# 'number' 값이 1이고, 'age'가 30 초과 40 미만인 행만 선택합니다.
df = df23.loc[df23['number'].isin([1]) & (df23['age'] > 30) & (df23['age'] < 40)]
# 필요한 컬럼('sex', 'number', 'age', 'income')만 선택하여 새로운 데이터프레임을 생성합니다.
df = df[['sex', 'number', 'age', 'income']]

# 새로운 데이터프레임의 마지막 3개 행을 출력합니다.
print(df.tail(3))
```

```python
##### KDE(커널 밀도 추정) 플롯 #####

# 성별이 남성(1)인 데이터만 선택하여 새로운 데이터프레임 생성
df_male = df.loc[df['sex'].isin([1])]
# 성별이 여성(2)인 데이터만 선택하여 새로운 데이터프레임 생성
df_female = df.loc[df['sex'].isin([2])]
# 남성의 소득에 대한 커널 밀도 추정 그래프를 그립니다.
sns.kdeplot(df_male['income'], shade=True, label='Male', clip=(-1000, 20000))
# 여성의 소득에 대한 커널 밀도 추정 그래프를 그립니다.
sns.kdeplot(df_female['income'], shade=True, label='Female', clip=(-1000, 20000))
# x축 레이블 설정
plt.xlabel("10,000won")
# 범례 표시
plt.legend()
# 격자 표시
plt.grid(True)

# 그래프를 화면에 출력
plt.show()
```

```python

selected_cols = [
    "조사연도", "수도권여부", "MD제공용_가구고유번호", "가구주_성별코드", 
    "가구주_만연령", "가구원수", "가구주_교육정도_학력코드", "가구주_혼인상태코드", 
    "자산", "부채", "처분가능소득(보완)[경상소득(보완)-비소비지출(보완)]", 
    "가구주_산업대분류코드", "가구주_직업대분류코드", "입주형태코드"
]

df23 = df_raw23[selected_cols]

# 열 이름 변경
df23.rename(columns={
    "조사연도": "year",
    "수도권여부": "metro",
    "MD제공용_가구고유번호": "id",
    "가구주_성별코드": "sex",
    "가구주_만연령": "age",
    "가구원수": "number",
    "가구주_교육정도_학력코드": "education",
    "가구주_혼인상태코드": "marriage",
    "자산": "asset",
    "부채": "debt",
    "처분가능소득(보완)[경상소득(보완)-비소비지출(보완)]": "income",
    "가구주_산업대분류코드": "industry",
    "가구주_직업대분류코드": "job",
    "입주형태코드": "house"
}, inplace=True)

print(df23.info())

# 조건에 맞는 데이터 선택
# 'number' 값이 1인 가구주, 'age'가 30 초과 40 미만인 가구주 데이터를 선택합니다.
df = df23.loc[
    df23['number'].isin([1]) &  # 'number' 값이 1인 가구주를 선택합니다. 'number' 값이 1인 경우는 1인 가구를 의미합니다.
    (df23['age'] > 30) &       # 'age'가 30을 초과하는 가구주를 선택합니다.
    (df23['age'] < 40)         # 'age'가 40 미만인 가구주를 선택합니다.
]

# 필요한 컬럼('sex', 'number', 'age', 'income')만 선택하여 새로운 데이터프레임을 생성합니다.
df = df[['sex', 'number', 'age', 'income']]

print(df.tail(3))
```

```python
import numpy as np
import pandas as pd


# 성별에 따른 데이터프레임 분리
df_male = df[df['sex'] == 1]  # 'sex' 값이 1인 행(남성)을 선택하여 새로운 데이터프레임 생성
df_female = df[df['sex'] == 2]  # 'sex' 값이 2인 행(여성)을 선택하여 새로운 데이터프레임 생성

# 평균 소득 계산
mn_male = df_male['income'].mean()  # 남성의 평균 소득 계산
mn_female = df_female['income'].mean()  # 여성의 평균 소득 계산
mn_diff = mn_male - mn_female  # 남성과 여성의 평균 소득 차이 계산
print("mn_diff:", mn_diff)

# 표준편차 계산
std_male = df_male['income'].std()  # 남성의 소득 표준 편차 계산
std_female = df_female['income'].std()  # 여성의 소득 표준 편차 계산

# 표준오차 계산
se_male = std_male / np.sqrt(len(df_male))  # 남성 소득의 표준 오차 계산
se_female = std_female / np.sqrt(len(df_female))  # 여성 소득의 표준 오차 계산
se_tot = np.sqrt(se_male**2 + se_female**2)  # 두 그룹의 표준 오차를 합산하여 전체 표준 오차 계산
print('표준오차:', np.round(se_tot, 3))

# Z-스코어 계산
z_score = mn_diff / se_tot  # 평균 소득 차이를 전체 표준 오차로 나누어 Z-스코어 계산
print("Z-스코어:", round(z_score, 3))
```

```python
import scipy as sp
import scipy.stats

# 정규분포를 따르는 연속 확률 변수 생성 (mu=0, var=1)
rv = sp.stats.norm(loc=0, scale=1)

# Z-스코어를 이용한 p-값 계산
p_value = 1 - rv.cdf(z_score)
print("유의 수준 (p-value):", np.round(p_value, 5))
```
